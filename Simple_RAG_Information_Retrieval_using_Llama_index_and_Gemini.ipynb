```python
from llama_index.llms.gemini import Gemini
from llama_index.embeddings.gemini import GeminiEmbedding
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, ServiceContext, StorageContext
import google.generativeai as genai
import os
```

```python
documents = SimpleDirectoryReader("/content/drive/MyDrive/Colab Notebooks/Information_Retrieval(RAG)_using_llama_index_and_gemini/")
documents = documents.load_data()
```

```python
print(documents[0].text)
```

```python
model = Gemini(model="models/gemini-1.5-flash")
```

```python
embedded_model = GeminiEmbedding(model_name="models/embedding-001")
```

```python
from llama_index.core import Settings

Settings.llm = model
Settings.embed_model = embedded_model
Settings.chunk_size = 800
Settings.overlap = 200
```

```python
index = VectorStoreIndex.from_documents(documents)
```

```python
index.storage_context.persist()
```

```python
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("What is ML?")
print(response)
```
