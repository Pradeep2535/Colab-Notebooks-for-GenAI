```python
def vector_search(query, collection):
  query_embeddings = get_embedding(query)

  if(not query_embeddings):
    return "Invalid query or embedding failed"
  pipeline = [
    {
       "$vectorSearch": {
            "index": "vector_index",
            "queryVector": query_embedding,
            "path": "embeddings",
            "numCandidates": 150,  # Number of candidate matches to consider
            "limit": 4,  # Return top 4 matches
        }
    },
    {
        "$project" : {
            "fullplot":1,
            "title":1,
            "genres":1,
            "score":{"$meta":"vectorSearchScore"},
        }
    }
  ]
  return list(collection.aggregate(pipeline))

```

```python
vector_search(user_query, collection)
```

```python
def get_search_result(query,collection):
  results = vector_search(query, collection)
  search_result = ""
  for result in results:
    search_result += f"Title : {result.get('title','N/A')},Plot: {result.get('fullplot', 'N/A')}\n"
  return search_result
```

```python
query = "recommend the best horror movie?"
```

```python
index_results = get_search_result(query, collection)
print(index_results)
```

```python
combined_information = f"Query: {query}\nContinue to answer the query by using the Search Results:\n{index_results}."
print(combined_information)
```

```python
from langchain.schema import Document
class CustomRetriever:
    def __init__(self, collection):
        self.collection = collection

    def get_relevant_documents(self, query):
        results = vector_search(query, self.collection)
        documents = []
        for result in results:
            documents.append(
                Document(page_content = result.get('fullplot', 'N/A'), metadata={"title": result.get('title', 'N/A')})

            )
        return documents

    def __call__(self, query):
        return self.get_relevant_documents(query)
```
