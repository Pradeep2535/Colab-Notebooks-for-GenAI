```python
query = "What is the best horror film?"
```

```python
from google.colab import userdata

PINECONE_API_KEY = userdata.get("PINECONE_API_KEY")
```

```python
from pinecone import Pinecone

pc = Pinecone(api_key=PINECONE_API_KEY)
index = pc.Index("mongodb-pinecone")
```

```python
def get_result(query, similar):
  vector=embedding_model.encode(query).tolist()
  result = index.query(
      vector=vector,
      top_k=similar,

  )
  return result
```

```python
results = get_result(query,3)
results
```

```python
result_matches = results['matches']
result_matches
```

```python
from bson import ObjectId
combined_information = ""
for i in range(len(result_matches)):
  obj = collection.find_one(ObjectId(result_matches[i]['id']))
  title = obj['title']
  fullplot=obj['fullplot']
  combined_information += f"Title : {title}\nFullPlot : {fullplot}\n\n"
combined_information
```

```python
prompt = f"Query: {query}\nContinue to answer the query by using the fullplot of the movies given only:\n{combined_information}."
```
